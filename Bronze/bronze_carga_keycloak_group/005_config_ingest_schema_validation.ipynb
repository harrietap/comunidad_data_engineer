```python
from pyspark.sql.types import StructType, StructField, StringType, BooleanType, TimestampType, IntegerType,ArrayType

schema_tbl_config_ingesta = StructType([
  StructField("id", IntegerType(), True),
  StructField("cliente", StringType(), True),
  StructField("fuente", StringType(), True),
  StructField("ruta_archivo", StringType(), True),
  StructField("formato", StringType(), True),
  StructField("tabla_destino", StringType(), True),
  StructField("frecuencia", StringType(), True),
  StructField("activo", BooleanType(), True),
  StructField("ultima_ejecucion", TimestampType(), True)
])

######
df_config_ingesta = spark.createDataFrame([], schema_tbl_config_ingesta)

df_config_ingesta.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze.tbl_config_ingesta")
```

```python
from pyspark.sql.types import StructType, StructField, StringType, BooleanType, TimestampType, IntegerType,ArrayType

schema_tbl_monitoreo_ingesta = StructType([
    StructField("id_ejecucion", IntegerType(), True),
    StructField("id_config", IntegerType(), True),
    StructField("fecha_inicio", TimestampType(), True),
    StructField("fecha_fin", TimestampType(), True),
    StructField("estado", StringType(), True),
    StructField("registros_cargados", ArrayType(
        StructType([
            StructField("archivo", StringType(), True),
            StructField("registros", IntegerType(), True)
        ])
    ), True),

    StructField("error_mensaje", StringType(), True)
])


df_monitoreo_ingesta = spark.createDataFrame([], schema_tbl_monitoreo_ingesta)

df_monitoreo_ingesta.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze.tbl_monitoreo_ingesta")
```
